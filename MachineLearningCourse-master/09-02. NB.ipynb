# Наивный байсовский классификатор

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import mysk

X, y = mysk.generate_points(seed = 0)
mysk.draw_points(X, y)
mysk.draw_centers()
```

```python
mysk.draw_points(X, y)
mysk.draw_centers()
mysk.draw_bayes_sep_curve()
```

```python
X_test, y_test = mysk.generate_points(seed = 1)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb_model = GaussianNB()
gnb_model.fit(X, y)
```

```python
#from sklearn.naive_bayes import BernoulliNB
#gnb_model = BernoulliNB(binarize = 0.1)
#gnb_model.fit(X, y)
```

```python
mysk.draw_points(X, y)
mysk.draw_sep_curve(gnb_model)
mysk.draw_bayes()
plt.scatter(gnb_model.theta_[:, 0], gnb_model.theta_[:, 1], color = ['b', 'r'], s = 100)
```

```python
gnb_model.theta_ # Средние для каждого класса (В LDA они назывались means_)
```

```python
err_train = np.mean(y != gnb_model.predict(X))
err_test = np.mean(y_test != gnb_model.predict(X_test))
print err_train, err_test
```
